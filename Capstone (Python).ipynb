```python
df=pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
```

```python
df
```

```python
def extract_names(column_value):
    genre_list = json.loads(column_value)
    genre_names = [item['name'] for item in genre_list]
    return ', '.join(genre_names)
```

```python
l1=['genres','keywords','production_companies','production_countries','spoken_languages']
```

```python
for i in l1:
    df[i] = df[i].map(extract_names)
```

```python
print("Number of Rows =",df.shape[0],"\n"
      "Number of Columns =",df.shape[1])
```

```python
print("Correlation between popularity and budget:",correlation)
```

correlation coefficient of 0.4319901360079756 suggests a moderate positive correlation between the budget of a movie and its popularity. Interpreting this correlation, we can say that, on average, as the movie budget increases, the movie's popularity tends to increase as well. This implies that higher-budget movies may have a greater chance of attracting a larger audience or generating more interest among viewers , but budget may not be the only essential element as other factors may also affect the popularity

#### Task 10

```python
filtered_df=df.sort_values(by='revenue', ascending=False).head(500)
filtered_df
```

```python
a=filtered_df[['budget','revenue','runtime']]
```

```python
a
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
pd.DataFrame({'Mean':a.mean(),'Median':a.median(),'Mode':a.mode().iloc[0]}).rename_axis('Measures')
```

```python
col=['budget','revenue','runtime']
```

```python
fig, ax= plt.subplots(1,1, figsize=(20,7))
filtered_df[col].boxplot()
```

```python
for i in col:
    Q1 = filtered_df[i].quantile(0.25)
    Q3 = filtered_df[i].quantile(0.75)
    IQR = Q3 - Q1
    outliers = filtered_df[(filtered_df[i] < Q1 - 1.5 * IQR) | (filtered_df[i] > Q3 + 1.5 * IQR)][i]
    filtered_df.drop(outliers.index, inplace=True)
```

```python
fig, ax= plt.subplots(1,1, figsize=(20,7))
filtered_df[col].boxplot()
```
